```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['feature1','feature2']].values # convert into an array for tensor flow

y=df['price'].values# convert into an array for tensor flow
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Normalizer
```

```python
help(MinMaxScaler)
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train1=scaler.transform(X_train)
```

```python
X_test1 =scaler.transform(X_test)
```

# Creating the model

```python
from tensorflow.keras.models import Sequential #Base sequentail model
from tensorflow.keras.layers import Dense #Keep adding layers to it
```

```python
#help(Sequential)
```

```python
#help(Dense)
```

```python
#Same model
model=Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
#model.add(Dense(4,activation='relu'))
#model.add(Dense(4,activation='relu'))

#Output Layer
model.add(Dense(1))

#Compiling the model, loss function
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train1,y=y_train,epochs=250)
```

```python
model.history.history['loss']
```

```python
loss_df=pd.DataFrame(model.history.history['loss'])
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test1,y_test)
```

```python
model.evaluate(X_train1,y_train)
```

```python
test_predictions=model.predict(X_test1)
```

```python
test_predictions=pd.Series(test_predictions.reshape(300,))
```

```python
test_predictions
```

```python
pred_df=pd.DataFrame(y_test,columns=['True Y'])
```

```python
pred_df
```

```python
pred_df['Predicted Y']=test_predictions
```

```python
pred_df
```

```python
sns.scatterplot(x='True Y',y='Predicted Y',data=pred_df)
```
