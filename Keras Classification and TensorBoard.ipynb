# 2nd Solution- Add Dropout layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model1 = Sequential()

model1.add(Dense(30,activation='relu'))
model1.add(Dropout(0.5))#rate= % of neurons turned of during each batch(Random Selection)

model1.add(Dense(15,activation='relu'))
model1.add(Dropout(0.5))
#Because its a BINARY CLASSIFICATION problem
model1.add(Dense(1,activation='sigmoid'))

model1.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25) # PATIENCE- Wait number of epochs after detection
```

```python
model1.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
losses2=pd.DataFrame(model1.history.history)
```

```python
losses2.plot()
```

```python
#Eventually flattening losses
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
predictions = model1.predict_classes(X_test) #For classsification 
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
print('\n')
print(accuracy_score(y_test,predictions))
print('\n')
```

```python
print('Accuracy= ',accuracy_score(y_test,predictions)*100,'%')
```
