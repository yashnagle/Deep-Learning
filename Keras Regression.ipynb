# Creating Model

```python
X = df.drop('price',axis=1).values
```

```python
X
```

```python
y=df['price'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))#last layer

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,batch_size=128, validation_data=(X_test,y_test),
          epochs=400)#batch size should be in powers of 2
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
#model.history.history
```

```python
pd.DataFrame(model.history.history)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score,accuracy_score
```

```python
predictions=model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test, predictions)
```
